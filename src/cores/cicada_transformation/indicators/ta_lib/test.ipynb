```python
import gcsfs
import pandas as pd

# Chemin complet vers le fichier GCS
full_path = 'gs://production-trustia-raw-data/Transformed/Crypto/binance/time_bars/BTCUSDT/1h/indicators/data_2024_09_20_2024_09_20'

# Créer un système de fichiers GCS
fs = gcsfs.GCSFileSystem()

# Utilisez fs.open pour ouvrir le fichier sur GCS
with fs.open(full_path) as f:
    df = pd.read_parquet(f)

# Afficher les premières lignes du DataFrame
print(df.head())
```
